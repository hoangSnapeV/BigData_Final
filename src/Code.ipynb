- Encode song -> time series represenation
- Save songs to text file. 

```python
def audio_to_midi(audio_file, midi_file):
   # Step 1: Load the audio file
   y, sr = librosa.load(audio_file)


   # Step 2: Extract pitch and timing information
   pitches, magnitudes = librosa.core.piptrack(y=y, sr=sr)


   # Step 3: Convert pitch and timing to MIDI events
   midi = pretty_midi.PrettyMIDI()
   instrument = pretty_midi.Instrument(0)


   for pitch, magnitudes in zip(pitches, magnitudes):
       # Choose the maximum magnitude as velocity
       velocity = int(max(magnitudes))
      
       # Skip zero frequencies to avoid divide by zero issues
       nonzero_pitches = pitch[pitch != 0]
       if len(nonzero_pitches) == 0:
           continue


       # Convert pitch to MIDI note
       note_number = int(librosa.hz_to_midi(np.mean(nonzero_pitches)))


       # Create a MIDI note event
       note = pretty_midi.Note(
           velocity=velocity,
           pitch=note_number,
           start=0,  # adjust as needed
           end=1    # adjust as needed
       )
       instrument.notes.append(note)


   # Step 4: Add the instrument to MIDI object
   midi.instruments.append(instrument)


   # Step 5: Save the MIDI file
   midi.write(midi_file)
```

```python
# Step 6: Specify the input audio file and output MIDI file
# path = "data/stimuli/test-stimuli-200-2009-05-29/"
path = "data/testData/"
audio_file_path = path + "002.wav"
midi_file_path = 'data/outputTest/004.mid'
```

```python
def concateListSongs(notes):
    listSongs = []
    for note in notes:
        songs = []
        # print(note[0])
        for element in note[0]:
            # print(element.pitch)
            songs.append(element.pitch.midi)
        listSongs.append(songs)
            
    return listSongs

# print(concateListSongs(notes))        
# print(len(concateListSongs(notes)))        
def mapping(listSongs):
    listMapping = []
    for song in listSongs:
        mapping = {}
        for i, symbol in enumerate(song):
            mapping[symbol] = i
            
        listMapping.append(mapping)
    return listMapping
```

```python
#create a reverse mapping 

def covert_songs_to_int(listSongs, map):
    intSongs = []
    for i, song in enumerate(listSongs):
        m = map[i]
        intAsong = []
        # print(type(song))
        # print(m)
        for note in song:
            intAsong.append(m[note])
        # print(song)
        intSongs.append(intAsong)
        
    return intSongs
```
